```python
from bs4 import BeautifulSoup
import urllib
import pandas as pd
from urllib.request import urlopen
from urllib.request import Request
from user_agent import generate_user_agent

url = 'https://www.chicagomag.com/chicago-magazine/november-2012/best-sandwiches-chicago/'

req = Request(url=url, headers={'User-Agent':generate_user_agent()})
page = urlopen(req)
soup = BeautifulSoup(page, "html.parser")
```

```python
item = soup.find_all('div', 'sammy')[0]
item.find(class_="sammyRank").string
```

```python
import re

text_list = re.split(('\n'), item.find(class_="sammyListing").get_text())

print(text_list[0])
print(text_list[1])
```

```python
from urllib.parse import urljoin

sql = """insert into sandwich values (%s, %s, %s, %s, %s)"""

for idx, item in enumerate(soup.find_all('div','sammy')):
    
    ranking = item.find(class_='sammyRank').string

    text_list = re.split(('\n'), item.find(class_="sammyListing").get_text())

    address = urljoin("https://www.chicagomag.com", item.find('a')['href'])

    req = Request(url=address, headers={'User-Agent':generate_user_agent()})
    page = urlopen(req)
    soup1 = BeautifulSoup(page, "html.parser")

    url_temp = soup1.find('p',class_="addy")
    text = url_temp.find('em').get_text()
    # print(text, end="\ ")
    re.split(".,", text)
    text = re.split(".,", text)[0]
    re.search("\$\d+.(\d+)?", text).group()
    texttemp = re.search("\$\d+.(\d+)?", text).group()
    price = float(texttemp.split("$")[1])

    address = urljoin("https://www.chicagomag.com", item.find('a')['href'])

    print(ranking, end=" ")
    print(text_list[0], end=" ")
    print(text_list[1], end=" ")
    print(price, end=" ")
    print(address)

    temp = tuple((ranking, text_list[1], text_list[0], price, address))
    cur.execute(sql, tuple(temp))
    conn.commit()
conn.close()
```
