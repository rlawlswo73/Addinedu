```python
import os

def modify_indices(directory):
    """
    Modify class indices in YOLO annotation .txt files.
    - Changes 0 -> 3
    - Changes 1 -> 6
    
    Parameters:
    - directory: Directory containing YOLO .txt annotation files.
    """
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            
            with open(file_path, "r") as file:
                lines = file.readlines()
            
            updated_lines = []
            for line in lines:
                parts = line.strip().split()
                if parts:  # Ensure the line is not empty
                    if parts[0] == "0":
                        parts[0] = "3"
                    elif parts[0] == "1":
                        parts[0] = "6"
                    updated_lines.append(" ".join(parts))
            
            # Save the updated content back to the file
            with open(file_path, "w") as file:
                file.write("\n".join(updated_lines))
            
            print(f"Updated indices in: {filename}")

# Example usage
directory_path = "/home/kjj37/dev_ws/yolo/yolov8/data/temp/valid/labels/"  # Update this to your folder path
modify_indices(directory_path)
```
