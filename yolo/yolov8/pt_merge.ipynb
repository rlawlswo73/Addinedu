```python
import os
import shutil

def organize_files_by_extension(folder_path):
    # .txt 파일용 폴더
    txt_folder = os.path.join(folder_path, "txt_files")
    other_folder = os.path.join(folder_path, "other_files")
    
    # 폴더 생성
    os.makedirs(txt_folder, exist_ok=True)
    os.makedirs(other_folder, exist_ok=True)
    
    # 폴더 내 파일 처리
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        
        # 파일인 경우에만 처리
        if os.path.isfile(file_path):
            if file_name.endswith(".txt"):
                shutil.move(file_path, os.path.join(txt_folder, file_name))
            else:
                shutil.move(file_path, os.path.join(other_folder, file_name))

# 사용 예시
folder_path = "/home/kjj37/dev_ws/yolo/yolov8/data/dataset5/images"  # 여기에 정리할 폴더 경로를 입력하세요.
organize_files_by_extension(folder_path)
```

```python
# import os

# os.environ["CUDA_DEVICE_ORDER"] = "0"
import torch

# GPU 상태 및 캐시 정리 함수
def clear_and_check_gpu_cache():
    if not torch.cuda.is_available():
        print("CUDA is not available. No cache to clear.")
        return

    # 현재 GPU 상태 출력
    print("Before clearing GPU cache:")
    print(torch.cuda.memory_summary())

    # 캐시 정리
    print("Clearing GPU cache...")
    torch.cuda.empty_cache()
    torch.cuda.synchronize()

    # 캐시 정리 후 상태 출력
    print("After clearing GPU cache:")
    print(torch.cuda.memory_summary())
```

```python
# 사용 예시
clear_and_check_gpu_cache()
import torch
import os

# GPU 사용 가능 여부 확인
if torch.cuda.is_available():
    print(f"GPU is available: {torch.cuda.get_device_name(0)}")
else:
    print("GPU is not available, using CPU.")

os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```
