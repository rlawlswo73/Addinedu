```python
import av
import os

#      
video_path = './data/videos/drive.mp4'
output_dir = './data/videos/drive_1s'
os.makedirs(output_dir, exist_ok=True)

#  
container = av.open(video_path)
video_stream = container.streams.video[0]

# FPS  1  
fps = video_stream.average_rate  #  FPS (: 60)
frame_interval = int(fps)  #  1 

frame_index = 0
saved_frame_count = 0

for frame in container.decode(video=0):
    if frame_index % frame_interval == 0:
        output_path = os.path.join(output_dir, f"frame_{saved_frame_count:06d}.jpg")
        frame.to_image().save(output_path)
        print(f"Saved: {output_path}")
        saved_frame_count += 1
    frame_index += 1

print(f"Total Frames Saved: {saved_frame_count}")
```
