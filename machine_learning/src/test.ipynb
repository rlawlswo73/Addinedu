```python
import pandas as pd
import numpy as np
```

```python
year = np.array([1983, 1984, 1985, 1986, 1987, 1988])
future_year = np.array([2015,2016,2017,2018,2019,2020])
future_avr = np.array([12000, 13000, 14000, 15000, 16000, 17000])
choi = np.array([2000, 2400, 3000, 3500, 3800, 4000])
k_lt = np.array([1200, 1500, 1800, 2000, 2200, 2500])
k_all = np.array([1200, 1500, 1800, 2000, 2200, 2500])
```

```python
from sklearn.linear_model import LinearRegression

choi_pd = pd.DataFrame(choi)
year_pd = pd.DataFrame(year)
X = pd.concat([choi_pd, year_pd])
y = year

reg = LinearRegression()
reg.fit(choi_pd, y)
```

```python
from sklearn.metrics import mean_squared_error

pred_tr = reg.predict(choi_pd)
pred_test = reg.predict(future_year.reshape(-1,1))

rmse_tr = (np.sqrt(mean_squared_error(y, pred_tr)))
rmse_test = (np.sqrt(mean_squared_error(y, pred_test)))

print("RMSE of Train Data : ", rmse_tr)
print("RMSE of Test Data : ", rmse_test)
```
